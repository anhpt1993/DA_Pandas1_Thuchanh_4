# 0. Import các thư viện

```python
import pandas as pd
```

# 1. Đọc dữ liệu từ file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2021/12/31021019/FoodPrice_in_Turkey.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# 2. Sử dụng phương thức iloc để truy cập dữ liệu

```python
# Truy cập dòng có chỉ số 3 của dữ liệu
tg = df.iloc[3]
tg
```

```python
# Truy cập các dòng có chỉ số liên tục từ 3 tới 7 của dữ liệu
tg = df.iloc[3: 8]
tg
```

```python
# Truy cập các dòng rời rạc của dữ liệu
tg = df.iloc[[3, 5, 9]]
tg
```

```python
# Truy cập cột thứ 4 của dữ liệu
# : được sử dụng để đại diện cho tất cả các dòng
# Cột có index = 5, cột "Month" sẽ được truy cập như sau

tg = df.iloc[:, 5]
tg.head()
```

```python
# Truy cập các cột liên tục từ cột 3 tới cột 7 của dữ liệu
tg = df.iloc[:,3:8]
tg.head()
```

```python
# Truy cập tới phần tử tại dòng 3, cột 7 của DataFrame
tg = df.iloc[3, 7]
tg
```

```python
# Truy cập tới các phần tử từ dòng 3 đến dòng 4, cột 5 đến cột 6 của dữ liệu
tg = df.iloc[3:5, 5:7]
tg
```

# 3. Sử dụng phương thức loc để truy cập dữ liệu. loc có thể làm việc được với chỉ số hàng, tên cột và các biểu thức điều kiện của cột

```python
# Truy cập dòng có chỉ số 3 của dữ liệu
tg = df.loc[3]
tg
```

```python
# Truy cập cột "ProductName" của dữ liệu
tg = df.loc[:,"ProductName"]
tg
```

```python
# Truy cập cột thứ 4 và 5 của dữ liệu
tg = df.loc[:,["UmName", "Month"]]
tg
```

```python
# Truy cập tới phần tử tại dòng 3, cột 7 của dữ liệu

tg = df.loc[3, "Price"]
tg
```

```python
# Truy cập tới phần tử có Year >= 2019
tg = df.loc[df.Year >= 2019]
tg
```

# 4. Thay thế các giá trị trong DataFrame

```python
# Thay số 5 bằng số 10 trong toàn bộ dữ liệu
df.replace(5, 10, inplace = True)
df.head()
```

```python
# Thay mã sản phẩm từ giá trị 52 thành RR trong toàn bộ dữ liệu
# RR là chuỗi ký tự nên được để trong cặp dấu nháy đơn

df.replace(52, "RR", inplace = True)
df.head()
```

```python
# Thay giá trị 10 trong cột "Month" thành giá trị 5
df["Month"].replace(10, 5, inplace = True)
df.head()
```
